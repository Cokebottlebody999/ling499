```python
import pandas as pd
```

```python
years = [1820, 1830, 1840, 1850, 1860, 1870, 1880, 1890, 1900, 1910, 1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000]
verbs = ['maintain', 'demand']
```

```python
for v in verbs:
    for y in years:
        print(v + ' '+ str(y))
```

```python
f = open("queryPartitionsPart2.sql", "a+")
for v in verbs:
    for y in years:
        sqlQuery = """
COPY (
SELECT lemma, count(*) from (
SELECT c.w2 FROM """ + 'p' + str(y) + """ as c, lexicon as l
where l.lemma like '""" + v + """'
and c.w1 = l.wid
UNION ALL
SELECT c.w3 FROM """ + 'p' + str(y) + """ as c, lexicon as l
where l.lemma like '""" + v + """'
and c.w1 = l.wid
UNION ALL
SELECT c.w4 FROM """ + 'p' + str(y) + """ as c, lexicon as l
where l.lemma like '""" + v + """'
and c.w1 = l.wid
UNION ALL
SELECT c.w5 FROM """ + 'p' + str(y) + """ as c, lexicon as l
where l.lemma like '""" + v + """'
and c.w1 = l.wid
UNION ALL
SELECT c.w6 FROM """ + 'p' + str(y) + """ as c, lexicon as l
where l.lemma like '""" + v + """'
and c.w1 = l.wid
) a, lexicon b where 
( b.pos like '%nn@%' or b.pos like '%nn\%%' or b.pos like 'nn' or b.pos like '%nn1%' or b.pos like '%nn2%')
and a.w2 = b.wid group by b.lemma order by count(*) desc
) TO """ + "\'/var/lib/postgresql/" + v + str(y) + ".csv\' CSV HEADER;\n"

        print(sqlQuery)
        f.write(sqlQuery)
f.close()
```
