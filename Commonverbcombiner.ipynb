```python
# years = [1820, 1830, 1840, 1850, 1860, 1870, 1880, 1890, 1900, 1910, 1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000]
years = [1830, 1840, 1850]
d = pd.DataFrame()
results = pd.DataFrame()
results['lemma'] = []
for i in years:
    print(i)
    file = "mostcommonverb3" + str(i) + ".csv"
    print(file)
    a = pd.read_csv(file)
    print(a.head())
    a['count']= a['count'].fillna(0)
    a[str(i)] = a['count']
    a = a.drop(["count"],axis=1)
    a = a.dropna(subset=['lemma'])
    results = results.merge(a, on = ["lemma"], how='outer')
results = results.fillna(0)
results["sum"] = results.sum(axis=1)
sort = results.sort_values(by=['sum'], ascending=False)
sort['lemma'] = '[' + sort['lemma'].str.upper() + ']'
sort
```

```python
sort.to_csv("mostcommonverbs.csv", encoding='utf-8', index=False)
```
