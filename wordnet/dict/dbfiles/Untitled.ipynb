```python
import re
import pandas as pd
```

```python
files = ["noun.act", "noun.cognition", "noun.communication","noun.event", "noun.feeling", "noun.location", "noun.motive", "noun.phenomenon", "noun.possession", "noun.process", "noun.relation", "noun.state"]
for f in files:
    print(f)
```

```python
wordsArray = []
i = -1
for file in files:
    with open(file) as f:
        for l in f:
            if "[" in l:
                p = re.compile('\[(.*)\]')
                matches = p.findall(l)
                for match in matches:
                    p2 = re.compile("([^,]+)")
                    matches2 = p2.findall(match)
                    for match2 in matches2:
                        result = re.sub(r'[0-9]+', '', match2)
                        result = re.sub(r'\+', '', result)
                        result = re.sub(r'\]', '', result)
                        result = re.sub(r'\[', '', result)
                        result = re.sub(r' ', '', result)
                        result = re.sub(r'!', '', result)
                        if "." not in result and result is not "":
                            i = i + 1
                            wordsArray.append([result])
                        elif "verb" in result and (result.split(':')[1] not in wordsArray[i] or len(wordsArray[i]) < 2):
                            wordsArray[i].append(result.split(':')[1])
newWordArray = []
for word in wordsArray:
    newWordArray.append(word)
```

```python
df = pd.DataFrame(newWordArray)
```

```python
df.to_csv("WordnetNouns.csv",  index = False)
```

```python
df = df.sort_values(by=[0])
```
