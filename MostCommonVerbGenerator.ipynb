```python
import pandas
```

```python
years = [1820, 1830, 1840, 1850, 1860, 1870, 1880, 1890, 1900, 1910, 1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000]
```

```python
f=open("mostcommonverb2.sql", "a+")
for i in years:
    a = pandas.read_csv("6M Partition/" + str(i) + ".csv")
    sqlSource = ""
    for tid in a["textID"]:
        sqlSource = sqlSource + "textid = " + str(tid) + " or "
    sqlSource = sqlSource[:-3]
    sqlQuery = "select lexicon.lemma, count(*) from corpus, lexicon where corpus.wid = lexicon.wid and "
    sqlQuery = sqlQuery + "(" + sqlSource + ")"
    sqlQuery = sqlQuery + "and lexicon.pos like '%vv%' and lexicon.pos not like '%nn%'"
    sqlQuery = sqlQuery + "group by lexicon.lemma order by count(*) DESC;"
    print(sqlQuery + "\n")
    f.write(sqlQuery)
    
f.close()
```
